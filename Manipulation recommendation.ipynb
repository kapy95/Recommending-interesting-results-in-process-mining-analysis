```python
import numpy as np
import pandas as pd
import pm4py
from scipy import stats
import import_ipynb
#from logEnrichment import enrichment
from interestingness_functions import compute_interestingness_kl_divergence,variance,schutzFormula,compute_interestingness_formula,support
```

```python
log_bpi_challenge=pd.read_csv("Log of analysis of time aspects.csv",index_col=0)
```

```python
dimensions=pd.read_excel("Operation dimensions.xlsx")
dimensions['Variables'] =  dimensions['Variables'].str.split().str.join(' ')
dimensions['Criteria']=dimensions['Criteria'].str.split().str.join(' ')
```

```python
operations_unique=log_bpi_challenge.drop_duplicates(subset=['Variant']).reset_index()
operations_unique=operations_unique[operations_unique['Type']=="Manipulation"]
```

```python
operations_unique
```

```python
types=[]
variables=[]
criteria=[]
variants=[]


for variant,optypes in zip(operations_unique['Variant'],operations_unique['Type']):
    row_variant=dimensions[dimensions['Variants']==variant]
    variants.append(variant)
    variables.append(row_variant["Variables"].values[0])
    criteria.append(row_variant["Criteria"].values[0])
    types.append(optypes)
```

```python
manipulation_dimensions=pd.DataFrame({"Variant":variants,"Types":types,"Variables":variables,"Criteria":criteria})
```

```python
manipulation_dimensions.to_csv("manipulationDimensions.csv")
```
